```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib

import glob
import json
```

```python
metadata_path = "/home/aaron/workspace/camhd_analysis/CamHD_motion_metadata/"

analysis_glob = metadata_path + "RS03ASHS/PN03B/06-CAMHDA301/2016/03/01/*_optical_flow.json"
print(analysis_glob)
```

```python
data = pd.DataFrame()

for filename in glob.glob(analysis_glob,recursive=True):
    print("Processing %s" % filename)
    
    with open( filename ) as file:
        j = json.load( file )
        timing = [frame["opticalFlow"]["performance"]["timing"] for frame in j["frameStats"] if frame["opticalFlow"]["valid"]]
        
        p = pd.DataFrame( timing )
        
        data = data.append(pd.DataFrame( timing ))
        
        
print(data)
```

```python
for key in ['minimization', 'opticalFlow',  'retrieveFrame1',  'retrieveFrame2']:
    pct_times[key] = data[key] / data['all'] * 100


print( pct_times )

print( pct_times.mean() )
print( data.mean() )
```
